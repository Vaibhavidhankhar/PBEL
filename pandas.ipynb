Pandas is an open-source Python library designed for data manipulation and analysis. It provides powerful and flexible data structures, primarily Series and DataFrames, to make working with "relational" or "labeled" data intuitive and efficient.
pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool,
built on top of the Python programming language.

Key Features and Uses:
- Data Structures:
Series: A one-dimensional labeled array capable of holding any data type (integers, strings, floating-point numbers, Python objects, etc.). It is essentially a column in a table.
- DataFrame:
A two-dimensional labeled data structure with columns of potentially different types, similar to a spreadsheet or SQL table. It is the most commonly used Pandas object.
Data Import/Export:
Pandas can read and write data from various formats, including CSV, Excel, JSON, SQL databases, and Parquet.
- Data Cleaning and Preparation:
It offers extensive functionalities for handling missing data (NaN values), removing duplicates, transforming data types, and reshaping data.
- Data Analysis and Manipulation:
Pandas provides tools for filtering, sorting, grouping, aggregating, merging, and joining datasets, enabling comprehensive statistical analysis and data exploration.
- Integration:
It integrates seamlessly with other Python libraries like NumPy for numerical operations and Matplotlib for data visualization.


Purpose:

The primary goal of Pandas is to be a fundamental high-level building block for practical, real-world data analysis in Python. It aims to simplify and accelerate tasks related to data cleaning, transformation, and analysis, making it an indispensable tool for data scientists, analysts, and researchers.

```python
# you hv to find out probability of winning if a team play in mumbai city
```

```python
# input for city as well as the team has to be taken from user


def get_team_win_counts(dataframe, team_name, specific_city=None):

    total_wins_all_cities = dataframe[dataframe['winner'] == team_name].shape[0]

    wins_in_specific_city = 0
    if specific_city:

        wins_in_specific_city = dataframe[
            (dataframe['winner'] == team_name) &
            (dataframe['city'].str.lower() == specific_city.lower())
            ].shape[0]

        return wins_in_specific_city, total_wins_all_cities

team_example = 'Chennai Super Kings'
city_example = 'Mumbai'


wins_in_city_count, total_wins_count = get_team_win_counts(df, team_example, city_example)

if total_wins_count is not None:
    print(f"\nResults for '{team_example}' :")
    print(f"Total times '{team_example}' won in {city_example} : {wins_in_city_count} times")
    print(f"Total times '{team_example}'won across all cities: {total_wins_count} times")
else:
    print(f"No data available for '{team_example}'. Please check team name or team data.")


print("\n" + "="*50 + "\n")

print("probability",wins_in_city_count/total_wins_count)
```
