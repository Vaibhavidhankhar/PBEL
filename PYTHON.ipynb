```python
hungry = input("Are you hungry? (yes/no): ").lower()

if hungry == "yes":
    time_of_day = input("Is it morning, afternoon, or evening? ").lower()

    if time_of_day == "morning":
        print("Let's make some breakfast!")
    elif time_of_day == "afternoon":
        print("Time for lunch!")
    elif time_of_day == "evening":
        print("What's for dinner?")
    else:
        print("Invalid time of day entered.")
elif hungry == "no":
    print("Okay, maybe later then.")
else:
    print("Please answer 'yes' or 'no'.")
```
